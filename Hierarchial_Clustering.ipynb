```
import pandas as pd
import matplotlib.pyplot as plt 
```

```
data= pd.read_csv('/content/Mall_Customers.csv')
```

```
data=data.iloc[:, [3,4]].values
```

```
import scipy.cluster.hierarchy as sch
dendrogram_plot=sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('clusters')
plt.ylabel('Euclidean distances')
plt.show()
```

#### Looking at the dendrogram, I can imply that the best selection for the number of cluster is 5

```
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hat=hc.fit_predict(data)
```

```
plt.scatter(data[y_hat==0, 0], data[y_hat==0,1], color='blue', label='Careless')
plt.scatter(data[y_hat==1, 0], data[y_hat==1, 1], color='red', label='Spendthrift')
plt.scatter(data[y_hat==2, 0], data[y_hat==2, 1], color='green', label='Cautious')
plt.scatter(data[y_hat==3, 0], data[y_hat==3, 1], color='pink', label='miser')
plt.scatter(data[y_hat==4, 0], data[y_hat==4, 1], color='yellow', label='Careful')
plt.xlabel('Spending Score')
plt.ylabel('Annual Income')
plt.title('Annual Income v/s Spending Score')
plt.legend()
plt.show()
```
